# Vision Disability

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from matplotlib import style

plt.style.use('seaborn-v0_8')

x = np.array([2016, 2017, 2018, 2019, 2020])
y = np.array([11563836, 12121301, 13190605, 12822788, 12766560])

def func(x, a, b):
    y = a*x + b
    return y

alpha = optimize.curve_fit(func, xdata = x, ydata = y)[0]
print(alpha)
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.title('Linear Approximation of Number of Vision Disabilities')
plt.xlabel('Years')
plt.ylabel('Number of Vision Disabilities')
plt.show()

value = alpha[0]*2021 + alpha[1]
print("")
print("Estimated Number in 2021: ", value)
value = alpha[0]*2022 + alpha[1]
print("Estimated Number in 2022: ", value)
value = alpha[0]*2023 + alpha[1]
print("Estimated Number in 2023: ", value)
```
