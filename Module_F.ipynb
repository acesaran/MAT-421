# Chapter 20.1 - Numerical Differentiation Problem Statement

A grid is a evenly spaced set of points over some domain, on some interval. The step size of such a grid is the distance between adjacent points on this grid. For example, if 5 was the distance between x0 and x1 on a grid, then 5 must be the distance between x1 and x2 on the same grid.

# Chapter 20.2 - Finite Difference Approximating Derivatives

```python
#Using taylor series, we can approximate the derivative using steps along a 'grid'.
#In this example, I am approximating -2cos(x) with its finite difference approximation 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
h = 0.1
x = np.arange(0, 2*np.pi, h) 
y = -2*np.cos(x) 

forward_diff = np.diff(y)/h 
x_diff = x[:-1:] 
exact_solution = 2*np.sin(x_diff) 

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

# 20.4 - Numerical Differentiation with Noise

```python
#We can also use noise to view a function as if it were slightly off from its pure function.
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
x = np.arange(0, 2*np.pi, 0.01) 
omega = 80
epsilon = 0.02

y = -np.sin(x) 
y_noise = y + epsilon*np.cos(omega*x)

plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = '-sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = '-sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
