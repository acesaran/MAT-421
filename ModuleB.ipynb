# Chapter 17.1 - Interpolation Problem Statement

An interpolation function is a function that, given a data set of independent and dependent values, can be used to find the value of a new x* estimated from the exiting data points.

This makes existing functions seem 'smoother' as the estimated data points will round out the graphs.

# Chapter 17.2 - Linear Interpolation

Linear Interpolation is a form of interpolation that assumes the estimated point to be found lies on a line between the nearest points to the left and right of it.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 1 ,2]
y = [4, 2, 1]

f = interp1d(x, y)
y_hat = f(.5)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Chapter 17.3 - Cubic Spline Interpolation

  Cubic Spline Interpolation is a form of interpolation that, instead of assuming a line that connects each point, all points are connected by a cubic polynomial. 
  
The interpolation function here is a set of piecewise cubic functions that form the entire graph.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [1, 3, 5]
y = [5, 1, 4]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(1, 5, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
