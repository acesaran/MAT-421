# 3.2 Continuity and Differentiation

```python
from sympy import limit, oo, Symbol
x = Symbol('x')
y = (2*x+1)/(3*x+1)
limit(y, x, oo)
```

```python
from sympy import *
import numpy as np
x = Symbol('x')
y = x**3 + ln(x)
yprime = y.diff(x)
yprime
```

```python
from sympy import symbols, cos, diff
a, b, c = symbols('a b c', real=True)
f = a*c + c**b + log(b)*a
print(diff(f, a))
```

# 3.3 Unconstrained Optimization

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,8,50)
f=np.log(2*x)
plt.plot(x,f,color=(1,0,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('ln(x)')
a=1.5
b=7.5
lam = .4
c = lam * a + (1-lam)*b
fa = np.log(2*a)
fb = np.log(2*b)
fc = np.log(2*c)
fcc = lam *fa+(1-lam)*fb

plt.plot([a,a], [0,fa])
plt.plot([b,b], [0,fb])
plt.plot([c,c], [0,fc])
plt.plot([c,c], [0, fcc])
plt.plot([a,b], [fa,fb])
```
