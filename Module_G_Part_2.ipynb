# 21.4 - Simpson's Rule

Simpson's Rule approximates the area underneath a curve by fitting a polynomial on each subinterval and integrating it accordingly.

```python
#The following code uses simpson's rule to approximate the integral from 0 to pi/2 of -sin(x) using 11 evenly spaced grid points.
import numpy as np

a = 0
b = np.pi/2
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = -np.sin(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = -1 - I_simp

print(I_simp)
print(err_simp)
```

# 21.5 - Computing Integrals in Python

```python
#We can also use the python libraries to approximate integrals in some of the ways we have learned in the past.
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0, np.pi, 0.01)
F_exact = -2*np.cos(x)
F_approx = cumtrapz(2*np.sin(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} 2sin(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```

```python
#Using the quad function, we can integrate from a to b of a function f.
from scipy.integrate import quad 

I_quad, est_err_quad = \
          quad(np.sin, np.pi, 2*np.pi)
print(I_quad)
err_quad = -2 - I_quad
print(est_err_quad, err_quad)
```
