# 22.1 - ODE Initial Value Problem Statement 

A differential equation is a relationship betwene a function f(x) and its independent variable (x), and an amount of its derivatives. An ordinary differential equation [ODE] is a differential equation where its indepdendent variable is in one dimension.
A partial diffential equation is where x is a vector containing indepdenent variables and the partial derivatives can be of any order with respect to any combination of its variables.

There are also two ways we can solve differential equations, getting a general solution and a particular solution. A general solution satisfies the differential equation while a particular solution satisfies the differential equation and known values of the solution.

# 22.3 - The Euler Method

```python
#This example approximates the solution to the differential equation df(t)/dt = e^-t with the inital condition of f0 = -1 using steps of .05
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%matplotlib inline

f = lambda t, s: np.exp(-t)
h = 0.05
t = np.arange(0, 1 + h, h) 
s0 = -1 

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
#This code is the same approximation as last example, but with smaller steps to get a better approximation. With such a small step, it is barely distinguishable from the exact graph.
h = 0.001
t = np.arange(0, 1 + h, h)
s0 = -1 

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
